```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder,LabelEncoder
```

```python
trans=[('ohe1',OneHotEncoder(),['outlook']),('ohe2',OneHotEncoder(),['temp']),('ohe3',OneHotEncoder(),['humidity']),('ohe4',OneHotEncoder(),['wind'])]
```

```python
df=df.drop(columns=['day'])
```

```python
df.columns
```

```python
c_t=ColumnTransformer(transformers=trans,remainder='passthrough')
```

```python
new_df=c_t.fit_transform(df)
```

```python
new_df=pd.DataFrame(new_df,columns=c_t.get_feature_names_out())
```

```python
new_df
```

```python
new_df['remainder__play']=LabelEncoder().fit_transform(new_df['remainder__play'])
```
