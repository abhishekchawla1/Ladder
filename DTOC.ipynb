```python
from sklearn.cluster import DBSCAN

e=[0.1,1,2,5]
min=[3,5,10,15]

for i in e:
    for j in min:
        d=DBSCAN(eps=i,min_samples=j)
        d.fit(X)
        labels=d.labels_
        ax=plt.scatter(X[:,0],X[:,1],c=labels)
        plt.show()
```

```python
best_params={'eps':0.1,'min_samples':15}
```

```python
db=DBSCAN(**best_params)
```

```python
db.fit(X)
```

```python
db.labels_
```

```python
plt.scatter(X[:,0],X[:,1],c=db.labels_)
```
